```python
#import necessary library
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
#read image of type jpg
image=cv2.imread('/content/input.jpg',0)
```

```python
plt.imshow(image,cmap='gray')
```

```python
# Noise removal with Gaussian filter
image_blur=cv2.GaussianBlur(image,(3,3),0)
```

```python
#Display the edges using the Kenny method
image_canny=cv2.Canny(image_blur,threshold1=100,threshold2=120)
```

```python
plt.imshow(image_canny,cmap='gray')
```
