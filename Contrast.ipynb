**3. If the image is in color, convert it to grayscale, and then find the contrast of the image
( 255 - pixel intensity).**

```python
image=cv.imread('/content/drive/MyDrive/dataset/images/b.jpg')
img_grey=cv.cvtColor(image,cv.COLOR_RGB2GRAY)
contrast=cv.subtract(255,img_grey)
plt.figure(figsize=(10,18))

plt.subplot(121);plt.imshow(image[:,:,::-1]),plt.title('input');
plt.subplot(122);plt.imshow(contrast,cmap='gray'),plt.title('contrast');
```

**4. If the image is in gray, convert it to binary, and then find the contrast of the image
(255 - pixel intensity).**

```python
# Load our image as greyscale 
img = cv.imread('/content/drive/MyDrive/dataset/images/c.jpg', 0)

rec,thersh1=cv.threshold(img,127,255,cv.THRESH_BINARY)
contrast=cv.subtract(255,thersh1)
plt.figure(figsize=[10,5])
plt.subplot(121);plt.imshow(img, cmap='gray');plt.title("Original");
plt.subplot(122);plt.imshow(contrast, cmap='gray');plt.title("contrast");
```

**5. If the image is binary, find the contrast of the image (255 - pixel intensity).**

```python
img=cv.imread('/content/drive/MyDrive/dataset/images/a.jpg')
img2=img[:,:,1]
contrast=cv.subtract(255,img2)
plt.figure(figsize=([10,5]))
plt.subplot(121);plt.imshow(img);plt.title('orginal')
plt.subplot(122);plt.imshow(contrast,cmap='gray');plt.title('contrast')
```
